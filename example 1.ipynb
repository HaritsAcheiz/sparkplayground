```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Local-Jupyter-Spark") \
    .master("spark://172.18.0.2:7077") \
    .config("spark.jars", "shared-data/postgresql-42.7.5.jar") \
    .getOrCreate()
```

```python
df = spark.read \
    .format("jdbc") \
    .option("url", "jdbc:postgresql://172.18.0.3/testing") \
    .option("dbtable", "employees") \
    .option("user", "postgresuser") \
    .option("password", "postgrespass") \
    .option("driver", "org.postgresql.Driver") \
    .load()

df.show()
```
